```python
# 1. Negative profit
neg_profit = df[df['profit'] < 0]

# 2. Very low sales
low_sales = df[df['sales'] <= 1]

# 3. Very high sales
high_sales = df[df['sales'] > 20000]

# 4. High discount
high_discount = df[df['discount'] > 0.7]

# 5. Abnormal profit margin
abnormal_margin = df[(df['profit_margin'] < 0) | (df['profit_margin'] > 1)]

# 6. Postal code too low
low_postal = df[df['postal_code'] < 1000]

# 7. Postal code too high
high_postal = df[df['postal_code'] > 99999]

# --- Printing results ---
print("Negative Profit:\n", neg_profit, "\n")
print("Low Sales (<=1):\n", low_sales, "\n")
print("High Sales (>20000):\n", high_sales, "\n")
print("High Discount (>0.7):\n", high_discount, "\n")
print("Abnormal Profit Margin:\n", abnormal_margin, "\n")
print("Low Postal Code (<1000):\n", low_postal, "\n")
print("High Postal Code (>99999):\n", high_postal, "\n")
```

```python
issues_summary = {
    "Negative Profit": (df['profit'] < 0).sum(),
    "Low Sales (<=1)": (df['sales'] <= 1).sum(),
    "High Sales (>20000)": (df['sales'] > 20000).sum(),
    "High Discount (>0.7)": (df['discount'] > 0.7).sum(),
    "Abnormal Profit Margin (<0 or >1)": ((df['profit_margin'] < 0) | (df['profit_margin'] > 1)).sum(),
    "Low Postal Code (<1000)": (df['postal_code'] < 1000).sum(),
    "High Postal Code (>99999)": (df['postal_code'] > 99999).sum()
}

summary_df = pd.DataFrame(list(issues_summary.items()), columns=["Issue", "Count"])
print(summary_df)
```

```python
# 1. Remove negative profit rows (optional: you could also keep but flag them)
df = df[df['profit'] >= 0]

# 2. Remove or fix extreme sales values
df = df[(df['sales'] > 1) & (df['sales'] <= 20000)]

# 3. Remove rows with very high discount
df = df[df['discount'] <= 0.7]

# 4. Fix abnormal profit margins
df = df[(df['profit_margin'] >= 0) & (df['profit_margin'] <= 1)]

# 5. Remove rows with invalid postal codes
df = df[(df['postal_code'] >= 1000) & (df['postal_code'] <= 99999)]

# Final check after cleaning
print("Rows after cleaning:", len(df))
```

```python
# 1. Convert order_date and ship_date to datetime
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])

# 2. Create 'order_year' and 'order_month'
df['order_year'] = df['order_date'].dt.year
df['order_month'] = df['order_date'].dt.month

# 3. Calculate shipping days
df['shipping_days'] = (df['ship_date'] - df['order_date']).dt.days

# 4. Create a sales category column
df['sales_category'] = pd.cut(
    df['sales'],
    bins=[0, 500, 2000, float('inf')],
    labels=['Low', 'Medium', 'High']
)

# 5. Calculate profit per unit
df['profit_per_unit'] = df['profit'] / df['quantity']
```
